```python
def load_all_to_df(*datasets):
    """
    Input : strings of CSVs titles
    Output: pandas dataframe containg all records in all CSVs
    """
    df_list = [pd.read_csv(dataset) for dataset in datasets]
    return pd.concat(df_list, ignore_index=True)
```

```python
def get_links(name: str):
    gn = GoogleNews(region='US')
    gn.set_time_range('03/01/2019','03/01/2021')
    gn.search(name+"%2B politician")

    config = Config()
    config.browser_user_agent = USER_AGENT
    config.request_timeout = 10
    
    links = gn.get_links()
    print(name+' Download Begin')
    stories = [] 
    for url in links:
        article = Article(url, config=config)
        try:
            article.download()
            article.parse()

            title = article.title
            date = article.publish_date
            text = article.text

            article.nlp()
            keywords = article.keywords
            keywords.sort()

            stories.append([article.url,keywords,article.summary])
        except:
            pass

        
    print(name+' Download end')
    return stories
```

```python
df_members['query_results'] = df_members.apply(lambda x: get_links(x['name']), axis = 1)
```

```python
def save_links(names, count):
    for name in names:
        goog = GoogleNews(lang='en', region='US')
        goog.set_time_range('03/01/2019','03/01/2021')
        goog.search(name+"%2B politician")
        links = goog.get_links()
        filepath = f'Links2/pol_{count}.txt'.format(count=count)
        try:
            with open(filepath, 'w') as f:
                f.write('\n'.join(links))
            print(r'{count} | {name} has been saved succesfully'.format(name=name, count=count))
        except:
            print(r'{count} | {name} HAD MISTAKES AAAHHHH'.format(name=name, count=count))
            pass
        count += 1
```
